```python
from graphviz import Digraph
from graphviz import Graph
```

```python
# Create a directed graph with three nodes: A → B → C
```

```python
dot=Digraph()

dot.node('A')
dot.node('B')
dot.node('C')

dot.edge('A','B')
dot.edge('B','C')

dot
```

```python
# Create an undirected graph with nodes: X, Y, Z, W connected in a cycle (X—Y—Z—W—X).
```

```python
dot=Graph()

dot.node('X')
dot.node('Y')
dot.node('Z')
dot.node('W')

dot.edge('X','Y')
dot.edge('Y','Z')
dot.edge('Z','W')
dot.edge('W','X')

dot
```

```python
# Create a graph with the following:

#     Start node (ellipse)
#     Process node (box)
#     Decision node (diamond)
#     End node (ellipse)
```

```python
dot=Graph()

dot.node("Start",shape="ellipse")
dot.node("process",shape="box")
dot.node("Decision",shape="diamond")
dot.node("End",shape="ellipse")

dot.edge("Start","process")
dot.edge("process","Decision")
dot.edge("Decision","End")
dot
```

```python
# Styled Nodes and Edges:

#     Create a graph where:
#         Nodes are filled with different colors.
#         Edges have different styles (solid, dashed, bold).
#         Nodes have custom font sizes
```

```python
dot=Digraph()

dot.node("A",style="filled",fillcolor="red",fontsize="20pt")
dot.node("B",style="filled",fillcolor="green",fontsize="40pt")
dot.node("C",style="filled",fillcolor="yellow",fontsize="5pt")

dot.edge("A","B",style="dotted")
dot.edge("B","C",style="bold")
dot.edge("C","A",style="solid")



dot
```

```python
# Clusters:

#     Create two clusters (Cluster 1 and Cluster 2).

#     Place A, B in Cluster 1, and X, Y in Cluster 2.

#     Connect A → X with a directed edge.
```

```python
dot = Digraph()

with dot.subgraph() as s:
    s.node("A")
    s.node("B")

with dot.subgraph() as s1:
    s1.node("X")
    s1.node("Y")
    


dot.edge("A","X")
dot
```

```python
# Subgraphs for Layered Layout:

#     Create a subgraph where nodes P1, P2 are at the same level.

#     Nodes M1, M2 are at another level.

#     Connect them in a meaningful way.
```

```python
dot = Digraph()

with dot.subgraph(name="Cluster_1") as P1:
    P1.attr(label="Cluster 1")
    P1.node("P1")
    P1.node("P2")
    P1.edge("P1","P2")



dot.node("M1")
dot.node("M2")

dot.edge("M1","P1")


dot
```

```python
# Complex Flowchart:

#     Create a flowchart:

#         Start → Process 1

#         Process 1 → Decision?

#         Decision? → End (if "Yes")

#         Decision? → Process 1 (if "No")
```

```python
dot= Digraph()

dot.node("Start")
dot.node("Decision ?")


with dot.subgraph() as s:
    s.node("Process 1")
    s.node("End")


dot.edge("Start","Process 1")
dot.edge("Process 1","Decision ?")
dot.edge("Decision ?","End",label="Yes")
dot.edge("Decision ?","Process 1",label="No",style="dotted")

dot
```

```python
# Tree Structure:

#     Create a binary tree with depth 3.

#     Each parent node should have two children.
```

```python
dot = Digraph()

dot.attr(rankdir='TB')  # Top to Bottom (default)
dot.attr('node', shape='circle', style='filled', fillcolor='lightblue')

dot.node('A','root')


dot.node('B')
dot.node('C')

dot.node('D')
dot.node('E')

dot.node('F',)
dot.node('G')

dot.edge('A','B')
dot.edge('A','C')

dot.edge('B','D')
dot.edge('B','E')

dot.edge('C','F')
dot.edge('C','G')
dot
```
