```python
from graphviz import Digraph
from graphviz import Graph
```

```python
Create a directed graph with three nodes: A → B → C
```

```python
dot=Digraph()

dot.node('A')
dot.node('B')
dot.node('C')

dot.edge('A','B')
dot.edge('B','C')

dot
```

```python
Create an undirected graph with nodes: X, Y, Z, W connected in a cycle (X—Y—Z—W—X).
```

```python
dot=Graph()

dot.node('X')
dot.node('Y')
dot.node('Z')
dot.node('W')

dot.edge('X','Y')
dot.edge('Y','Z')
dot.edge('Z','W')
dot.edge('W','X')

dot
```

```python
Create a graph with the following:

    Start node (ellipse)
    Process node (box)
    Decision node (diamond)
    End node (ellipse)
```

```python
dot=Graph()

dot.node("Start",shape="ellipse")
dot.node("process",shape="box")
dot.node("Decision",shape="diamond")
dot.node("End",shape="ellipse")

dot.edge("Start","process")
dot.edge("process","Decision")
dot.edge("Decision","End")
dot
```

```python
Edge Labels:

    Create a directed graph with three nodes (A, B, C).
    Label the edges:
        A → B with "Step 1"
        B → C with "Step 2"
```

```python
dot= Digraph()

dot.node("A")
dot.node("B")
dot.node("C")

dot.edge("A","B",label="Step 1")
dot.edge("B","C",label="Step 2")
dot
```
