```python
# Clusters:

#     Create two clusters (Cluster 1 and Cluster 2).

#     Place A, B in Cluster 1, and X, Y in Cluster 2.

#     Connect A → X with a directed edge.
```

```python
dot = Digraph()

with dot.subgraph() as s:
    s.node("A")
    s.node("B")

with dot.subgraph() as s1:
    s1.node("X")
    s1.node("Y")
    


dot.edge("A","X")
dot
```

```python
# Subgraphs for Layered Layout:

#     Create a subgraph where nodes P1, P2 are at the same level.

#     Nodes M1, M2 are at another level.

#     Connect them in a meaningful way.
```

```python
dot = Digraph()

with dot.subgraph(name="Cluster_1") as P:
    P.attr(label="Cluster 1")
    P.node("P1")
    P.node("P2")
    P.edge("P1","P2")



dot.node("M1")
dot.node("M2")

dot.edge("M1","P1")


dot
```

```python
# Tree Structure:

#     Create a binary tree with depth 3.

#     Each parent node should have two children.
```

```python
dot = Digraph()

dot.attr(rankdir='TB')  # Top to Bottom (default)
dot.attr('node', shape='circle', style='filled', fillcolor='lightblue')

dot.node('A','root')


dot.node('B')
dot.node('C')

dot.node('D')
dot.node('E')

dot.node('F',)
dot.node('G')

dot.edge('A','B')
dot.edge('A','C')

dot.edge('B','D')
dot.edge('B','E')

dot.edge('C','F')
dot.edge('C','G')
dot
```

```python
# Network Graph:

#     Represent a network with:

#         3 servers (S1, S2, S3)

#         2 clients (C1, C2)

#         A firewall node in between
```

```python
dot=Graph()

dot.attr(rankdir="LR")

with dot.subgraph(name="Server") as s:
    s.attr(label="Server")
    s.node("S1")
    s.node("S2")
    s.node("S3")

with dot.subgraph(name="Client") as c:
    c.attr(label="Label")
    c.node("C1")
    c.node("C2")

dot.node("FireWall")

dot.edge("C1","FireWall")
dot.edge("C2","FireWall")
dot.edge("FireWall","S1")
dot.edge("FireWall","S2")
dot.edge("FireWall","S3")

dot
```

```python
# AI Neural Network Representation:

#     Create a neural network structure:

#         Input layer: 3 neurons

#         Hidden layer: 4 neurons

#         Output layer: 1 neuron

#         Connect all layers properly.
```
